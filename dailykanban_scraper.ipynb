```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_page(url):
    i = 1
    while True:
        new_url = url.replace('1', str(i))
        r = requests.get(new_url)
        if r.status_code != 404:
            yield r.text
            if i % 10 == 0 or i == 1:
                print('Yield Page %d.' %i)
            i += 1
        else:
            print("Return 404.")
            break
```

```python
def update_list(page, title_list, url_list):
    tag_soup = BeautifulSoup(page, 'html.parser')
    for link_code in tag_soup.find_all('h2', 'entry-title'):
        article_title = link_code.get_text()
        article_url = link_code.a.get('href')
        
        title_list.append(article_title)
        url_list.append(article_url)
```

```python
def update_df(df, url):
    title_list = []
    url_list = []
    
    for page in get_page(url):
        update_list(page, title_list, url_list)

    df['title'] = title_list
    df['url'] = url_list
```

```python
def output(df, filename):
    df.to_csv(filename + '.csv')
    print('Compelete.')
```
